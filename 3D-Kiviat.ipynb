# 3D-Kiviat on a Hydraulic dataset

## Definition of the data

```python
#%matplotlib inline
from kiviat import (Kiviat3D, Tree)

plabels = ['Ks1', 'Ks2', 'Ks3', 'Q']
bounds = [[30., 30., 30., 3000.], [40., 40., 40., 7000.]]

sample = np.loadtxt('data/sample_mascaret.dat')
data = np.loadtxt('data/data_mascaret.dat')

print(f'Sample shape {sample.shape} -> data shape {data.shape}')
```

## Kiviat

```python
# Only one sample
kiviat = Kiviat3D(sample[None, 0], data[0, None], bounds=bounds, plabels=plabels)
fig, ax = kiviat.plot(flabel='Water level (m)')
```

```python
# Every sample
kiviat = Kiviat3D(sample, data, bounds=bounds,
                  plabels=plabels, idx=245,
                  stack_order='qoi',
                  cbar_order='hdr')
kiviat.plot(flabel='PDF')
```

```python
kiviat = Kiviat3D(sample, data, bounds=bounds,
                  plabels=plabels)
kiviat.plot(flabel='Water level (m)')
```

```python
# Functional Hypothetical Outcome plot
kiviat.f_hops(frame_rate=400)
```

## Tree plot

```python
# Tree
tree = Tree(sample[:, [0, 3]], data, bounds=[[30., 3000.], [40., 7000.]],
            plabels=['Ks3', 'Q'])
tree.plot(flabel='Water level (m)')
```

```python
# Functional Hypothetical Outcome plot
tree.f_hops(frame_rate=400)
```
